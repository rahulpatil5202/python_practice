```python
import re
import string
```

```python
with open('./rahulsoutlook/Sent.csv', mode='rb') as f:  # The 'b' in mode changes the open() function to read out bytes.
    bytes = f.read()
    text = bytes.decode('utf-8', 'ignore') # Change 'ignore' to 'replace' to insert a '?' whenever it finds an unknown byte.
```

```python
match = re.findall(r'[\w\.-]+@[\w\.-]+\.\w+', text)
match
```

```python
with open('rahul_sent.txt', 'w') as f:
    for item in match:
        f.write("%s\n" % item)
```
